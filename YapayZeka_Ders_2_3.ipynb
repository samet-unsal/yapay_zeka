```python
# 1. Aşama: Görüntüyü yükleme

image = cv2.imread('Opencv_resimler/siberakademi_banner.png') # 'your_image.jpg' yerine görüntü dosya
```

```python
# 2. Görüntüyü görüntüleme

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Yüklenen Görünütünün Gösterimi")
plt.show()
```

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap = 'gray')
plt.title("Gri Tonlanmış Görüntü")
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(gray_image,(5,5),0)
plt.imshow(blurred_image,cmap='gray')
plt.title("Bulanıklaştırılmış Görüntü")
plt.show()
```

```python
edges = cv2.Canny(blurred_image, threshold1=100, threshold2=100)
plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama Sonucu")
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,4,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("Orijinal Görüntü")

plt.subplot(1,4,2)
plt.imshow(gray_image,cmap='gray')
plt.title("Gri Tonlama")

plt.subplot(1,4,3)
plt.imshow(blurred_image, cmap = 'gray')
plt.title("Bulanık Görüntü")

plt.subplot(1,4,4)
plt.imshow(edges, cmap = 'gray')
plt.title("Kenar Algılama")

plt.show()
```

```python
image1 = cv2.imread('Opencv_resimler/trafik-1.png')
image2 = cv2.imread('Opencv_resimler/trafik-2.png')

image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

plt.subplot(1,2,1)
plt.imshow(image1)
plt.title("Görüntü 1")

plt.subplot(1,2,2)
plt.imshow(image2)
plt.title("Görüntü 2")

plt.show()
```

```python
image1_resized = cv2.resize(image1, (image2.shape[1], image2.shape[0]))

plt.subplot(1, 2, 1)
plt.imshow(image1_resized)
height, width, channels = image1_resized.shape
plt.title('Yeniden Boyutlandırılan 1. Görsel')
print("1. Görüntü genişliği (x ekseni):", width, "piksel", "1. Görüntü yüksekliği (y ekseni):", height, "piksel")

plt.subplot(1, 2, 2)
plt.imshow(image2)
height2, width2, channels2 = image2.shape
plt.title('Yeniden Boyutlandırılan 2. Görsel')
print("2. Görüntü genişliği (x ekseni):", width2, "piksel", "2. Görüntü yüksekliği (y ekseni):", height2, "piksel")

plt.show()
```

```python
import cv2

# Resmi yükle

img = cv2.imread('Opencv_resimler/trafik-1.png')

# Gri tonlamaya çevir

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# Otsu yöntemi ile otomatik eşik bulma
ret, thresh_img = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

print(f'Otsu ile bulunan eşik degeri: {ret}')
```

```python
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Orijinal Resim')
plt.axis('off') # eksenleri kaldırır
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret, thresh_img = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 2)
plt.imshow(thresh_img, cmap='gray')
plt.title("Otsu çıktısı: {}".format(ret))
plt.axis('off')

plt.show()
```
