```python
# Ali'nin ders ve not bilgileri
data = {'Ders':['Algoritma', 'BilgisayarProgramlama', 'Elektronik-2', 'AviyonSistemler'],
         'Not': [95, 100, 65 ,90]}

# Veriyi bir dataframe e dönüştürür.
df = pd.DataFrame(data)

# En yüksek notu bul
en_yuksek_not = df['Not'].max()
print("En yüksek not: ",en_yuksek_not)

# Ortalama notu hesapla
ortalama_not = df['Not'].mean()
print("Ortalama not: ",ortalama_not)


print(df.describe())
```

```python
import numpy as np

# Rastgele tanımlanan değerler - sırası olmayan
dizi = np.array([1,-55,9,-23,13,-21,23,37,29,-6])

# Dizinin ortalama sonuc
ortalama = np.mean(dizi)
print("Ortalama: ", ortalama)

# Dizinin standart sapması
std_sapma = np.std(dizi)
print("Standart sapma: ", std_sapma)

# Dizinin mutlak değerleri
mutlak_veriler = np.abs(dizi)
print("Mutlak Deger Ciktisi: ",mutlak_veriler)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

x = np.linspace(-5,5,100)
y = np.linspace(-5,10,50)
X,Y = np.meshgrid(x,y)
Z= np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X,Y,Z, cmap='plasma')
ax.set_title('3D Yüzey Grafiği Örneği')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_colormaps():
    colormaps = plt.colormaps()
    n = len(colormaps)
    fig, axs = plt.subplots(n//2, 2, figsize=(10, 2*(n//2)))
    for ax, cmap_name in zip(axs.ravel(), colormaps):
        cmap = plt.get_cmap(cmap_name)
        img = np.linspace(0, 1, 256).reshape(1, -1)
        ax.imshow(img, aspect='auto', cmap=cmap)
        ax.set_title(cmap_name, fontsize=8)
        ax.set_axis_off()
    plt.show()

plot_colormaps()
```

```python
import requests
from bs4 import BeautifulSoup

# Web sayasfının URL'si
url = 'https://gelisim.edu.tr/tr/gelisim-anasayfa'

# Web sayfasını indir
response = requests.get(url)

# eğer bağlantı başarılıysa
if response.status_code == 200:
  #HTML içeriğini ayrıştır
  soup = BeautifulSoup(response.content, 'html.parser')

  # Örnek olarak sayfadaki tüm başlıkları (h1,h2,h3,vb) çek
  for heading in soup.find_all(['h1','h2','h3']):
    print(heading.text.strip())
else:
  print("Bağlantı başarısız:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://alicetinkaya.site/'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    links = soup.find_all('a')

    for link in links:
      href = link.get('href')
      text = link.text.strip()
      if href:
        print("Bağlantı metni: ", text, "URL: ", href)
    print("Toplam bulunan bilgi sayısı: ", len(links)," adet")
else:
  print("Bağlantı başarısız: ", response.status_code)
```
