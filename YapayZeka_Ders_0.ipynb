```python
import pandas as pd
```

```python
# Ali'nin ders ve not bilgileri
data = {'Ders':['Algoritma', 'BilgisayarProgramlama', 'Elektronik-2', 'AviyonSistemler'],
         'Not': [95, 100, 65 ,90]}

# Veriyi bir dataframe e dönüştürür.
df = pd.DataFrame(data)

# En yüksek notu bul
en_yuksek_not = df['Not'].max()
print("En yüksek not: ",en_yuksek_not)

# Ortalama notu hesapla
ortalama_not = df['Not'].mean()
print("Ortalama not: ",ortalama_not)


print(df.describe())
```

```python
import numpy as np

# Rastgele tanımlanan değerler - sırası olmayan
dizi = np.array([1,-55,9,-23,13,-21,23,37,29,-6])

# Dizinin ortalama sonuc
ortalama = np.mean(dizi)
print("Ortalama: ", ortalama)

# Dizinin standart sapması
std_sapma = np.std(dizi)
print("Standart sapma: ", std_sapma)

# Dizinin mutlak değerleri
mutlak_veriler = np.abs(dizi)
print("Mutlak Deger Ciktisi: ",mutlak_veriler)
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x,y)
plt.title('Sinüs Grafiği')
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')
plt.grid(True) # Ekranı kare kare bölüyor
plt.show()
```

```python
# Yıllara göre satışlar
yillar = [2018, 2019, 2020, 2021, 2022]
satislar = [150, 200, 20, -225, 14]

# Satış grafiğini çiz
plt.plot(yillar,satislar,marker = 'o')

# Grafik başlıkları
plt.title("Yıllara Göre Satışlar")
plt.xlabel("Yıllar")
plt.ylabel("Satış Miktarı")

plt.xticks(yillar) # aralıkları direkt yıllara göre ayarlıyor.
# Grafiği göster
plt.grid()
plt.show()
```
